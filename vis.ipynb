```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pos_path = 'pos.csv'
dns_path = 'dns.csv'
```

```python
#pos_data = np.genfromtxt(pos_path, delimiter=',')
dns_data = np.genfromtxt(dns_path, delimiter=',')

nt = dns_data.shape[0]
ncells = dns_data.shape[1] - 1
print(f"nt={nt}, ncells={ncells}")
```

```python
plt.plot(dns_data.T);
```

```python
from IPython.display import HTML
from IPython.display import display

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()

ax.set_xlim((0,ncells))
ax.set_ylim((0,2))

line, = ax.plot([],[])

def update(frame):
    line.set_data(np.arange(ncells + 1), dns_data[frame])
    return line,

animation = FuncAnimation(fig, update, frames=range(nt), blit=True)

display(HTML(animation.to_jshtml()))
```
